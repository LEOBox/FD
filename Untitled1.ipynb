```python
from generator import Generator
from ssd300 import SSD300
from boxcoder import create_prior_boxes
from ssd_training import MultiboxLoss
from ssd_utils import BBoxUtility
```

```python
classes = ["Face"]
NUMCLASSES = len(classes) + 1
inputShape = (300, 300, 3)

model = SSD300(inputShape, numClasses=NUMCLASSES)

loss = MultiboxLoss(NUMCLASSES, neg_pos_ratio=2.0).compute_loss
model.compile(optimizer='Adadelta', loss=loss)
```

```python
priorBoxes = create_prior_boxes(model)
boxCoder = BBoxUtility(NUMCLASSES, priorBoxes)
```

```python
priorBoxes
```

```python
import numpy as np
import pickle
data = pickle.load(open("annotations.pickle",'rb'))
for key in data:
    for d in data[key]:
        d.pop(-2)
    data[key] = np.array(data[key])
keys = list(data.keys())
train_num = int(0.7 * len(keys))
train_keys = keys[:train_num]
val_keys = keys[train_num:]

gen = Generator(gt=data, bbox_util=boxCoder,
                 batch_size=16, path_prefix='./processed/',
                 train_keys=train_keys, val_keys=val_keys, image_size=(300, 300))
```

```python
len(keys)
```

```python
epochs = 5
model.fit_generator(generator=gen.generate(True), steps_per_epoch=int(gen.train_batches / 16),
                              validation_data=gen.generate(False), validation_steps=int(gen.val_batches / 16),
                              epochs=epochs, verbose=1)
#model.save_weights("face")
```

```python
model.load_weights("face_scipy")
```

```python
from keras.preprocessing import image
from scipy.misc import imread
from keras.applications.imagenet_utils import preprocess_input
import numpy as np

files = ('sample.jpg','sample1.jpg','sample3.jpeg','sample4.jpg','sample5.jpg','sample6.jpeg','sample7.jpg','sample8.jpg',
        'sample9.jpg','sample10.jpg')

inputs = []
images = []
for f in files:
    f = "./{}".format(f)
    img = image.load_img(""+f, target_size=(300, 300))
    img = image.img_to_array(img)
    
    images.append(imread(f))
    inputs.append(img.copy())

inputs = preprocess_input(np.array(inputs))

preds = model.predict(inputs, batch_size=1, verbose=1)
# preds.shape (5, 7308, 33)
results = boxCoder.detection_out(preds)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

for i, img in enumerate(images):
    # Parse the outputs.
    det_label = results[i][:, 0]
    det_conf = results[i][:, 1]
    det_xmin = results[i][:, 2]
    det_ymin = results[i][:, 3]
    det_xmax = results[i][:, 4]
    det_ymax = results[i][:, 5]

    # Get detections with confidence higher than 0.6.
    top_indices = [i for i, conf in enumerate(det_conf) if conf >= 0.5]

    top_conf = det_conf[top_indices]
    top_label_indices = det_label[top_indices].tolist()
    top_xmin = det_xmin[top_indices]
    top_ymin = det_ymin[top_indices]
    top_xmax = det_xmax[top_indices]
    top_ymax = det_ymax[top_indices]

    colors = plt.cm.hsv(np.linspace(0, 1, 21)).tolist()

    plt.imshow(img / 255.)
    currentAxis = plt.gca()

    for i in range(top_conf.shape[0]):
        xmin = int(round(top_xmin[i] * img.shape[1]))
        ymin = int(round(top_ymin[i] * img.shape[0]))
        xmax = int(round(top_xmax[i] * img.shape[1]))
        ymax = int(round(top_ymax[i] * img.shape[0]))
        score = top_conf[i]
        label = int(top_label_indices[i])
        label_name = classes[label - 1]
        display_txt = '{:0.2f}, {}'.format(score, label_name)
        coords = (xmin, ymin), xmax-xmin+1, ymax-ymin+1
        color = colors[label]
        currentAxis.add_patch(plt.Rectangle(*coords, fill=False, edgecolor=color, linewidth=2))
        currentAxis.text(xmin, ymin, display_txt, bbox={'facecolor':color, 'alpha':0.5})
    
    plt.show()
```
